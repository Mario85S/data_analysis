```python
# Import pandas
import pandas as pd 



```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', delim_whitespace=True)
```

```python
# Compare the average income based on ethnicity and print the results
avg_income_by_ethnicity = df.groupby('Ethnicity')['Income'].mean()
print("Average income by ethnicity: ")
print(avg_income_by_ethnicity)
```

```python
# filter the married and single people into two separate DataFrames
married = df[df['Married'] == 'Yes']
single = df[df['Married'] == 'No']

# calculate the mean balance for both married and single people
married_bal_mean = married['Balance'].mean()
single_bal_mean = single['Balance'].mean() 

# compare the means and print the results
if married_bal_mean > single_bal_mean:
    print("Married people have a higher balance which on average is: ", married_bal_mean)
elif married_bal_mean < single_bal_mean:
    print("Single people have a higher balance which on average is: ", single_bal_mean)
else:
    print("Married and single people have the same average balance.")


```

```python
# Find highest income in balance dataset and print the results
highest_income = df.loc[:, "Income"].max()
print("The highest income in dataset is: ", highest_income)
```

```python
# Find lowest income in balance dataset and print the results
lowest_income = df.loc[:, "Income"].min()
print("The lowest income in dataset is: ", lowest_income)
```

```python
# Sum of recorded Cards in balance dataset and print the results
sum_cards = df.loc[:, "Cards"].sum()
print(" The sum of cards recorded in dataset is: ", sum_cards)
```

```python
# Create a variables for both genders and count how many female and male are in the dataset
female = df[df['Gender'] == 'Female'].shape[0]
male = df[df['Gender'] == 'Male'].shape[0]

# Print the results
print(f"There are {female} females and {male} males in the dataset.")


```
