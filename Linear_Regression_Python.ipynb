```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Get the current working directory
cwd = os.getcwd()

# Create the relative path to the file
file_path = os.path.join(cwd, r'C:\Users\mariu\OneDrive\Desktop\insurance.csv')

# Load the dataset
insurance_df = pd.read_csv(file_path, encoding='ascii')

# Display the head of the dataframe to ensure it's loaded correctly
insurance_df.head()
```

```python
# Extracting the relevant columns
X = insurance_df[['age']]
y = insurance_df['charges']

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.5)
plt.title('Insurance Charges vs Age')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True)
plt.show()
```

```python
# Fitting the linear regression model
model = LinearRegression()
model.fit(X, y)

# Making predictions
predictions = model.predict(X)
```

```python
# Plotting the scatter plot with the best-fit line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.5)
plt.plot(X, predictions, color='red', linewidth=2)
plt.title('Insurance Charges vs Age with Best-Fit Line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.grid(True)
plt.show()
```

This visualization illustrates the relationship between age and insurance charges, 
with the best-fit line indicating a general trend of increasing insurance charges with age.
